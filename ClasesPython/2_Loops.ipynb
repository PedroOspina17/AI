# For


Is controlled by the ***# of iterations***.

**structure: ``for i in range(1,10,1):``**

params on range function are assigned as following: **`range(from,to,increment)`**

```python
for i in range(1,11,1): # Last number (to value) is not included
    print(i)
```

```python
for i in range(0,10): # Increment could be ignored
    print(i)
print()    
for i in range(10): # value to start could be ignored. This creates a loop of 10 iterations
    print(i)
    
```

```python
start = int(input("Please type the start number"))
stop = int(input("Please type the stop number"))
increment = int(input("Please type the increment"))

for i in range(start,stop,increment):
    print(i)
```

```python
print(" \nBreaks the loop on iteration 5\n")
for i in range(0,10,1):    
    if(i == 5):
        break                     # STOPS THE LOOP
    print(i)
    
print(" \n Skips the iteration 5 \n")
for i in range(0,10,1):    
    if(i == 5):
        continue                    # JUMPS TO THE NEXT ITERATION
    print(i)
    
    
```

# While

Controled by a **Boolean** expression.

   >Be careful, you could create a infinite loop.

**Structure: ``while(Expression):``**



```python
nombre = "Pedro"
while(nombre=="Pedro"):
    print("Es pedro")
```

```python
nombre = "default"
while(nombre!=""):
    nombre = input("Please type your name: ")
    print("Hi {}".format(nombre))
```

You can use While to achieve the functionality of a for.

```python
i = 0
while(i<10):    
    print(i)
    i+=1
```

```python
i = 10
while(i>0):    
    print(i)
    i-=1
```
