# Amazon Web Services AWS

## Que es cloud computing?

Cloud Computing (computación en la nube) es la entrega a pedido (on-demand) de potencia informática, almacenamiento de bases de datos, aplicaciones y otros recursos de TI a través de una plataforma de servicios en la nube a través de Internet con precios de pago por uso (pay-as-you-go).

Lideres de Cloud Computing:
* Amazon Web Services (AWS)
* Google Cloud Platform (GCP)
* Microdoft - Azure Cloud Services

## AWS para Data Scientist

Herramientas que son utiles para los data scientist son las siguientes:

* **Elastic Compute Cloud (EC2)** : Son servidores virtuales que puede ejecutar cualquier programa o aplicacion.
* **Simple Storage Service (S3)** : Es un servicio de almacenamiento de archivos como Dropbox o Google Drive, pero más escalable y está hecho especialmente para trabajar con cóidigos y aplicaciones.
* **Relational Database Service (RDS)** : Es un servicio de base de datos relacional en la nube.  Es compatible con SQL Server, MySQL, PostgreSQL, Oracle, ect.

### Manipulación de Archivos y Permisos

`pwd` 

print working directory, permite conocer la ruta en la que estamos ubicados.

`ls directorio`

Lista el contenido de un directorio, si no se pone el directorio lista los archivos en el actual directorio.

`ls -a directorio`

Se puede usar -a para mostrar archivos ocultos.

`cd`

Change directory, es el comando que se requiere para acceder a una ruta diferente en la que se encuentra.

`cd ..`

Para subir de nivel, es decir que si estas en /home/profesor al ejecutar el comando pasarias a estar en /home

`mkdir`

Make directory, crea un directorio nuevo tomando en cuenta la ubicación actual.

`cp prueba.txt /home/respaldo/prueba.txt`

Copy, copia un archivo o directorio de un origen a un destino

`mv prueba.txt /home/respaldos/prueba2.txt`

Move, mueve un archivo a una ruta especifica

`rm /home/profesor/prueba.txt`

Remove, comando para eliminar un archivo o directorio

`rm -rf /home/respaldo`

Si se quiere eliminar un directorio, la r indica que elimina los archivos de forma recursiva y la f indica que borra sin pedir confirmación, este comando es delicado asi que debe estar seguro de ejecutarlo.

`rmdir`

Remove directory, elimina un directorio siempre que éste vacío.

`chown alumno prueba.txt`

Change owner, permite cambiar el propietario de un archivo o directorio.

`chown alumno prueba.txt`

Change mode, permite cambiar los permisos de un archivo o directorio.

`chown -R usuario:grupo /home/ejercicios`

La opcion -R indica que se realice de manera recursiva en todos los archivos en el directorio.

`chmod`

Change mode, permite cambiar los permisos de un archivo o directorio para los diferentes grupos de usuarios. 

Grupos de usuarios:

- u: usuario dueño del fichero
- g: grupo de usuarios asociado al fichero
- o: todos los otros usuarios
- a: todos los tipos de usuario (dueño, grupo y otros)

Los permisos disponibles son:

- 0 (cero) = Sin permisos
- 1 = permiso de ejecutar únicamente
- 2 = permiso de escribir solamente
- 3 = escribir y ejecutar
- 4 = permiso de leer
- 5 = leer y ejecutar
- 6 = leer y escribir
- 7 = leer, escribir y ejecutar

`chmod 444 hola.txt`

Permisos de solo lectura para el dueño, el grupo y todo los usuarios.

## SSH o Secure Shell

Es un protocolo de administración remota que permite a los usuarios controlar y modificar sus servidores remotos.

Los usuarios de Windows podrían aprovechar los clientes SSH como **Putty**.

La sintaxis de conexion es la siguiente:

`ssh {user}@{host}`

**Ejemplo con EC2**:

`ssh -i "keypar.pem" ec2-user@ID_INSTANCE.us-east-2.compute.amazonaws.com`

**Nota:** Usuarios Windows es con *Putty*

## Transferir archivos con SCP por SSH

Este comando está disponible tanto en Linux o Mac, para Windows podrias probar una herramienta que se llama **WinSCP** que permite realizar lo mismo con interfaz gráfica.

La sistaxis para subir un archivo desde tu maquina:

`scp nombre_archivo usuario@servidor:ruta_servidor_donde_colocar_archivo`

**Ejemplo con EC2**:

`scp -i keypar.pem ../DS4A_2020/week4/case_22.1/fetch_weather_data.py ec2-user@ID_INSTANCE.us-east-2.compute.amazonaws.com:/home/ec2-user/week4`

La sistaxis para descargar un archivo desde un servidor:

`scp usuario@servidor:ruta_servidor_donde_esta_el_archivo ruta_ordenador_local_donde_poner_el_archivo`

**Ejemplo con EC2**:

`scp -i keypar.pem ec2-user@ID_INSTANCE.us-east-2.compute.amazonaws.com:/home/ec2-user/week4/fetch_weather_data.py fetch_weather_data.py`

### AWS IAM : Control del acceso a recursos y servicios 

![](iam.jpg)

#### Los usuarios IAM

Son objetos de la cuenta que permiten a un usuario individual acceder al entorno AWS con un conjunto de credenciales. Los permisos se pueden aplicar individualmente a un usuario, pero no es lo recomendable, lo más idóneo es vincular cada usuario a un grupo IAM, y a este grupo sí que se le asignaran los permisos para acceder a los recursos y objetos de AWS.

#### Los grupos IAM

Son objetos que permiten gestionar de manera eficiente los permisos y el acceso a sus recursos en AWS. Es muy recomendable el uso de grupos IAM para gestionar los permisos y necesario en el caso de que haya un número considerable de usuarios con acceso a recursos AWS. 

#### Los roles IAM

Tienen similitudes a los grupos IAM aunque en lugar de proporcionar permisos a usuarios, los roles sirven para conceder permisos a las instancias en cuanto estás son creadas. De esta forma, las aplicaciones que corren en una instancia podrán utilizar estas credenciales para firmar las peticiones, por ejemplo, una aplicación que necesite tener acceso a un bucket de S3.

#### Políticas IAM

Las políticas IAM son un documento donde se enumeran las acciones que un usuario, grupo o rol puede realizar sobre los recursos de AWS. A continuación mencionamos algunas de las características de las políticas IAM:

* Son archivos en formato JSON.
* Por defecto el permiso está denegado (Deny by default)
* En el caso de que hayan varias políticas, las más restrictivas son las que prevalecen.
* El orden de prioridad es el siguiente:
 - Primero Explicit Deny
 - Segundo Explicit Allow
 - Tercero Implicit Deny
