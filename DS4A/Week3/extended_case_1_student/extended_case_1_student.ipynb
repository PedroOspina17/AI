### Exercise 8 (40 min):

Which hours have the most accidents for each borough? Plot a bar graph for each borough showing the number of accidents for each hour of the day.

**Hint:** You can use ```sns.FacetGrid``` to create a grid of plots with the hourly data of each borough. 

**Answer.**

````python
groupedData = monthlydf.groupby("BOROUGH")

fig = plt.figure(figsize=(15,8))

for i,(key,values) in enumerate(groupedData):
    
    
    values["HOUR"] = values.index.hour
    infoToPlot = values["HOUR"].value_counts()

    ax = plt.subplot(3,2,i+1,)
    plt.bar(infoToPlot.index,infoToPlot)
    plt.title("Accidents during the day in {}".format(key))
    plt.xlabel("Hour")
    plt.ylabel("Amount of accidents")
fig.tight_layout(pad=3.0)
#     _ = ax.xticks(infoToPlot.index)
````
