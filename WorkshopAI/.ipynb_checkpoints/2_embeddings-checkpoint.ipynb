# Embeddings

To embed means to reduce preserving the meaning.

```python
import json
import os
import sys
import pickle
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from sklearn.decomposition import PCA
import math
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.manifold import TSNE
```

Loading data preprocessed

```python
with open('1.2_subsampled_wikis.pickle', 'rb') as f:
    # Pickle the 'data' dictionary using the highest protocol available.
    (word2idx, idx2word, sentences) = pickle.load(f)
```

```python
V = len(word2idx)
N = len(sentences)
print(V, N)
```

```python
%matplotlib inline
def plot2D(X, Y):
    fig, ax = plt.subplots(1, 1, figsize=(15,15))
    plt.scatter(X, Y, ax=ax)
    for i in range(V):
        try:
            plt.annotate(s=idx2word[i].encode("utf8"), xy=(X[i], Y[i]))
        except:
            print ("bad string:", idx2word[i])
    plt.show()
    
def plot3D(X, Y, Z):
    fig = plt.figure()
    ax = Axes3D(fig)

    for i in range(V): #plot each point + it's index as text above
        ax.scatter(X[i], Y[i], Z[i], color='g')
        try:
            ax.text(X[i], Y[i], Z[i],  '%s' % (idx2word[i]), size=10, zorder=1,  color='b')
        except:
            print ("bad string:", idx2word[i])
    plt.show()
```

> **Exercise:** Create a term document matrix, named A, of size VxN, in which a count of tokens should be made per sentence (documents).

*Hint:* You can initialize your matrix with zeros using Numpy function zeros, like np.zeros((Rows, Columns))

```python
# create raw counts first
A = np.zeros((V, N))
j = 0
for sentence in sentences:
    for i in sentence:
        A[i,j] += 1
    j += 1
print ("finished getting raw counts")
```

Words to numbers

```python
transformer = TfidfTransformer()
A = transformer.fit_transform(A)
A = A.toarray()
```

Embedding...

```python
# Embedding data into 2 dimensions
pca = PCA(n_components=2)
We = pca.fit_transform(A)
```

```python
We
```

```python
# plot
plot2D(We[:,0], We[:,1])
```

```python
pca = PCA(n_components=3)
We = pca.fit_transform(A)
```

```python
plot3D(We[:,0], We[:,1], We[:,2])
```

> **Exercise:** Complete the function below. Add the sentence needed to calculate cosine distance, i.e. cosine similarity.

```python
def find_analogies(w1, w2, w3, We, word2idx, distance='Euclidean'):
    king = We[word2idx[w1]]
    man = We[word2idx[w2]]
    woman = We[word2idx[w3]]
    # HERE assing the proper operation for vector V0
    v0 = king - man + woman

    def dist_euclidean(a, b):
        return np.linalg.norm(a - b)

    # HERE create the cosine distance function named as dist_angular
    def dist_angular(a, b):
        return 1 - a.dot(b) / (np.linalg.norm(a) * np.linalg.norm(b))

    if distance == 'Euclidean':
        dist_func = dist_euclidean
    elif distance == 'Angular':
        dist_func = dist_angular
    else:
        print('create your custom function :P')
        
    min_dist = math.inf
    best_word = ''
    for word, idx in word2idx.items():
        if word not in (w1, w2, w3):
            v1 = We[idx]
            # CALCULATES DISTANCE BETWEEN TWO WORDS
            d = dist_func(v0, v1)
            if d < min_dist:
                min_dist = d
                best_word = word

    print("closest to " + w3 + ": " + best_word)
    print(w1, "-", w2, "=", best_word, "-", w3)
```

```python
find_analogies('king', 'man', 'family', We, word2idx, 'Angular')
# find_analogies('canada', 'canadian', 'argentina', We, word2idx)
# find_analogies('samsung', 'iphone', 'amazon', We, word2idx)
```
